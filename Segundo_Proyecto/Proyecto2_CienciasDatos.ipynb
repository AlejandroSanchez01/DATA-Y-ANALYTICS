```python
# convertimos en 1,2,3
paises=autos["Pais_Origen"].unique()
def pais(pais: str)-> int:
  """Devuelve el numero entero de acuerdo al pais
  Ejemplo
  'USA'    --> 1,
  'Europe' --> 2,
  'Japan'  --> 3
  """
  #Se recorre mediante un for la lista de paises para revisar a que posicion corresponde el pais
  #que se le va ingresar como parametro, y se devuelve i+1 para que empiece desde 1
  for i in range(len(paises)):
    if(pais==paises[i]):
       return i+1 
```

```python
autos["Pais_Origen"]=(autos["Pais_Origen"]).apply(pais)
```

- Generar 5 nuevas variables aplicando los siguientes criterios
con las variables numricas:
1. Log10(var1/var2)
2. Sqrt(var1)*exp(var2)/200
3. Si var 1 &gt; var 2  5, caso contrario 3
4. 1/logn(var1/var2)*100
5. var2**2/var1

```python
#Se agrega la columna "Log10(Desplazamiento/Aceleracion)"" en donde se aplica
#el logartimo en base 10 de "Desplazamiento/Aceleracion" con la funcion apply
autos["Log10(Desplazamiento/Aceleracion)"]=round((autos["Desplazamiento"]/autos["Aceleracion"]).apply(lambda x: math.log10(x)),2).astype('float')
```

```python
#Se agrega la columna aplica "Sqrt(Caballo_De_Fuerza)*exp(Aceleracion)/200" en donde 
#se aplica la raiz cuadrada a la variable Caballo_De_Fuerza, luego se multiplica por 
#el exponente de la variable Aceleracion y este se divide entre 200
autos["Sqrt(Caballo_De_Fuerza)*exp(Aceleracion)/200"]=round((
    (autos["Caballo_De_Fuerza"].apply(lambda x: math.sqrt(x)))
    *(autos["Aceleracion"].apply(lambda x: math.exp(x)))/200),2).astype('float')
```

```python
autos.reset_index(inplace=True, drop=False)
```

```python
#Se inserta la columna "Millas_Por_Galon>Aceleracion" si "Millas_Por_Galon" > "Aceleracion"  5, caso contrario 3
autos["Millas_Por_Galon>Aceleracion"] = (autos["Millas_Por_Galon"]>autos["Aceleracion"]).apply(lambda x: 
                                                    5 if x==True else (3
                                                    )
)
```

```python
#Se agrega la columna "1/logn(Peso_En_Libras/Caballo_De_Fuerza)*100"
autos["1/ln(Peso_En_Libras/Caballo_De_Fuerza)*100"] = round(1/(autos["Peso_En_Libras"]/autos["Caballo_De_Fuerza"]).apply(lambda x: math.log(x))*100,2).astype('float')
```

```python
#Se agrega por ultimo la columna "Cilindros"**2/"Millas_Por_Galon"
autos["Cilindros**2/Millas_Por_Galon"] = round(autos["Cilindros"]**2/autos["Millas_Por_Galon"],2).astype('float')
```

- Generar 3 variables cualitativas ordinales a partir de las
variables numricas con los siguientes criterios:
- Si es mayor a la media + 1*desviacin estndar 
Alto

- Si est entre media - 1*desviacin estndar y
media + 1*desviacin estndar  Medio

- Si es menor a la media - 1*desviacin estndar 
Bajo
Nota: Mostrar con las nuevas variables generadas.

**Elegir con libertad las variables a trabajar.**


```python
def mean(dataset: float)-> float:
    """Funcion que realiza la media 
    de un conjunto de datos
    """
    return sum(dataset) / len(dataset)
```

```python
import statistics

#Primera Variable
Alto = mean(autos["Log10(Desplazamiento/Aceleracion)"])+(1*statistics.pstdev(autos["Log10(Desplazamiento/Aceleracion)"]))
Bajo =  mean(autos["Log10(Desplazamiento/Aceleracion)"])-(1*statistics.pstdev(autos["Log10(Desplazamiento/Aceleracion)"]))
autos["Log10(Desplazamiento/Aceleracion)_CualitativaOrdinal"] = autos["Log10(Desplazamiento/Aceleracion)"].apply(
                                                                                          lambda x:
                                                                                          "Alto" if(x>Alto)
                                                                                          else "Medio" if (x>=Bajo and x<=Alto)
                                                                                          else "Bajo" if(x<Bajo) else "Sin Clasificar")
```

```python
#Segunda Variable
Alto = mean(autos["Sqrt(Caballo_De_Fuerza)*exp(Aceleracion)/200"])+(1*statistics.pstdev(autos["Sqrt(Caballo_De_Fuerza)*exp(Aceleracion)/200"]))
Bajo =  mean(autos["Sqrt(Caballo_De_Fuerza)*exp(Aceleracion)/200"])-(1*statistics.pstdev(autos["Sqrt(Caballo_De_Fuerza)*exp(Aceleracion)/200"]))

autos["Sqrt(Caballo_De_Fuerza)*exp(Aceleracion)/200_CualitativaOrdinal"] = autos["Sqrt(Caballo_De_Fuerza)*exp(Aceleracion)/200"].apply(
                                                                                          lambda x:
                                                                                          "Alto" if(x>Alto)
                                                                                          else "Medio" if (x>=Bajo and x<=Alto)
                                                                                          else "Bajo" if(x<Bajo) else "Sin Clasificar")
```

```python
#Tercera Variable
Alto = mean(autos["1/ln(Peso_En_Libras/Caballo_De_Fuerza)*100"])+(1*statistics.pstdev(autos["1/ln(Peso_En_Libras/Caballo_De_Fuerza)*100"]))
Bajo =  mean(autos["1/ln(Peso_En_Libras/Caballo_De_Fuerza)*100"])-(1*statistics.pstdev(autos["1/ln(Peso_En_Libras/Caballo_De_Fuerza)*100"]))

autos["1/ln(Peso_En_Libras/Caballo_De_Fuerza)*100_CualitativaOrdinal"] = autos["1/ln(Peso_En_Libras/Caballo_De_Fuerza)*100"].apply(
                                                                                          lambda x:
                                                                                          "Alto" if(x>Alto)
                                                                                          else "Medio" if (x>=Bajo and x<=Alto)
                                                                                          else "Bajo" if(x<Bajo) else "Sin Clasificar")
```

```python
# autos.Pais_Origen.unique()[0] -> USA
# autos.Pais_Origen.unique()[1] -> Japan
# autos.Pais_Origen.unique()[2] -> Europe

columnsUsa=list()
columnsJapan=list()
columnsEurope=list()
def agregarIndicesPais():
  fila=0
  for i in range(len(autos)):
    if(autos["Pais_Origen"].loc[i]==autos.Pais_Origen.unique()[0]):
      #Columnas que corresponden al pais USA
      columnsUsa.append(fila)
    elif(autos["Pais_Origen"].loc[i]==autos.Pais_Origen.unique()[1]):
      #Columnas que corresponden al pais JAPAN
      columnsJapan.append(fila)
    else:
      #Columnas que corresponden al pais EUROPE
      columnsEurope.append(fila)
    fila=fila+1

agregarIndicesPais()
```

```python
def mapa_calor(columna, pais):
  # correlacin con mapa de calor
  fig, ax = plt.subplots(tight_layout = True)
  hist = ax.hist2d(autos[columna], autos[pais])
  #Mayor frecuencia, mayor intensidad del color
```

```python
#Se establecen nuevos dataframes para poder realizar la matriz de correlacion por paises
autosUsa = autos.loc[columnsUsa,:]
autosJapan = autos.loc[columnsJapan,:]
autosEurope = autos.loc[columnsEurope,:]
```

```python
#Se genera la matriz de correlacin para todas las variables de tipo int y float
#Las columnas que tienen mayor correlacin se acercan al valor 1
def graficarMatrizCorrelacion(dataFrame,title):
  plt.figure(figsize=(8,6))
  correlacion = dataFrame.corr() # pearson
  ax = plt.axes()
  sns.heatmap(correlacion,
              ax=ax,
              xticklabels = correlacion.columns.values,
              yticklabels = correlacion.columns.values,
              annot = True,
              cmap = "viridis"  # viridis, Blues    
              );
  ax.set_title(title)           
  
graficarMatrizCorrelacion(autosUsa,"Pais_Origen (USA)")
graficarMatrizCorrelacion(autosJapan,"Pais_Origen (JAPAN)")
graficarMatrizCorrelacion(autosEurope,"Pais_Origen (EUROPE)")
```
